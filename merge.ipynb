```python
# read data.json into dataframe score
score = pd.read_json('data.json')
```

```python
score.head()
```

```python
# rename column 'url' to 'Domain name' in score dataframe
score.rename(columns={'url': 'Domain name'}, inplace=True)
score.head()
```

```python
# merge domain into score dataframe using 'Domain name' as key
merged = pd.merge(score, domain, on='Domain name')
```

```python
merged.head()  
```

```python
# save merged dataframe to csv
merged.to_csv('merged.csv', index=False)
```

```python
# show the column names of the merged dataframe
merged.columns
```

```python
# scored = merged delete columns 'status', 'redirect', 'dotgov','https', 'responseTime', 'keywords', 'ogImageWidth', 'ogImageHeight','twitterSite', 'twitterCard', 'twitterTitle','twitterDescription', 'twitterImage', 'history','Security contact email'
scored = merged.drop(['redirect', 'dotgov','https', 'responseTime', 'keywords', 'ogImageWidth', 'ogImageHeight','twitterSite', 'twitterCard', 'twitterTitle','twitterDescription', 'twitterImage', 'history','Security contact email'], axis=1)
# filter out rows with 'status' != '200'
scored = scored[scored.status == 200]
# scored drop column 'status'
scored = scored.drop(['status'], axis=1)
scored.to_csv('scored.csv', index=False)
```

```python
# for each value in scored, change "True" to 1 and "False" to 0
scored = scored.replace(True, 1)
scored = scored.replace(False, 0)
```

```python
scored.to_csv('scored.csv', index=False)
```

```python
# from column 3 to 15, sum all the values and save to a new column 'score'
scored['count'] = scored.iloc[:, 2:15].sum(axis=1)
# round x to 0 decimal places
scored['score'] = round(scored['count']/13*100)
```

```python
scored.head()
```

```python
scored.to_csv('scored.csv', index=False)
```

```python
# scored, group by 'State', and calculate the mean of column 'score'
state_mean = scored.groupby('State')['score'].mean().round()
# save state_mean to csv
state_mean.to_csv('state_mean.csv')
```

```python
# count how many True values in column title,description,robots,viewport,canonical,ogLocale,ogSiteName,ogType,ogTitle,ogDescription,ogUrl,ogImage,ogImageAlt
tags = scored[['title','description','robots','viewport','canonical','ogLocale','ogSiteName','ogType','ogTitle','ogDescription','ogUrl','ogImage','ogImageAlt']].apply(pd.Series.value_counts)
# for each column, divide the number by 852
tags = tags/852
tags.to_csv('tags.csv')
```

```python
# scored, group by 'State', and calculate the mean of column 'score'
tag_list = ['title','description','robots','viewport','canonical','ogLocale','ogSiteName','ogType','ogTitle','ogDescription','ogUrl','ogImage','ogImageAlt']
type_mean = scored.groupby('Domain type')[tag_list].mean()
# save state_mean to csv
type_mean.to_csv('type_mean.csv')
type_mean
```
